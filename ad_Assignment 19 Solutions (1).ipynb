**1.Create a function that takes a string and returns a string in which
each character is repeated once.**

**Examples:**  
double_char("String") ➞ "SSttrriinngg"  
double_char("Hello World!") ➞ "HHeelllloo WWoorrlldd!!"  
doublechar("1234!\_") ➞ "11223344!!\_\_"

**def** double_char(in_string):

out_string **=** ''

**for** ele **in** in_string:

out_string **+=** ele**\***2

**return** out_string

print(f'➞ {double_char("String")}')

print(f'➞ {double_char("Hello World!")}')

print(f'➞ {double_char("1234!\_")}')
