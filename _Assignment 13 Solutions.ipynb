**1. What advantages do Excel spreadsheets have over CSV spreadsheets?**

**Ans:** The Advantages of Excel over CSV are:

1.  Excel (XLS and XLSX) file formats are better for storing and
    analysing complex data.

2.  An Excel not only stores data but can also do operations on the data
    using macros, formulas etc

3.  CSV files are plain-text files, Does not contain formatting,
    formulas, macros, etc. It is also known as flat files

**Ans:** For csv.reader(iterable_file_object), the file objects needed
to be opened in read mode mode='r' Whereas

opened in write mode mode='w'
